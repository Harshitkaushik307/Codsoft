IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
```

DISPLAYING DATA

```python
data = pd.read_csv("codsoft_1.csv")
data.head()
```

DATA CLEANING

```python
data.nunique()
```

EXPLORING CABIN COLUMN

```python
cols = data.columns
print(cols)
data["Cabin"] = data["Cabin"].apply(lambda x: "X" if pd.isnull(x) else x[0])
data.head()
```

```python
x_labels = data["Cabin"].unique()
height_labels = []
for i in x_labels:
    height_labels.append(data["Cabin"][(data["Survived"] == 1)&(data["Cabin"] == i)].count())
height_labels = np.array(height_labels)
norm_val = np.linalg.norm(height_labels)
height_labels = height_labels/norm_val
print(height_labels)

plt.figure(figsize=(6,3))
plt.bar(x_labels,height_labels)
plt.show()
```

```python
data["Cabin_marked"] = data["Cabin"].apply(lambda x: 0 if x == "X" else 1)
data.drop(["Cabin","PassengerId","Name","Ticket"],axis=1,inplace=True)
```

IMPUTATIONS

```python

mean_age = data["Age"].mean()
data["Age"].fillna(mean_age,inplace=True)
data.dropna(inplace=True)

plt.figure(figsize=(6,3))
sns.heatmap(data.isnull())
plt.show()
```

ONE HOT ENCODING

```python
data = pd.get_dummies(data)
data.head()
data.drop(["Sex_female","Embarked_S"],axis=1,inplace=True)
```

SPLITTING DATA INTO X AND Y

```python
cols = data.columns
features,output = list(cols[1:]),[cols[0]]
print(features,output)
data.head()
```

EXPLORATORY DATA ANALYSIS

```python
it = 1
plt.figure(figsize=(6,6))

for i in cols:
    plt.subplot(6,3,it)
    plt.xlabel(i)
    plt.subplots_adjust(right=3,top=3)
    sns.kdeplot(data[i])
    plt.grid()
    it+=1

plt.show()
    
```

```python
plt.figure(figsize=(6,3))
sns.heatmap(data.corr())
plt.show()
```

SPLITTING DATA AND TRAINING LOGISTIC REGRESSION MODEL

```python
X,y = data.drop(["Survived"],axis = 1),data["Survived"]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
model = LogisticRegression(max_iter = 500)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

MODEL EVALUATION

```python
y_pred = model.predict(x_test)
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
